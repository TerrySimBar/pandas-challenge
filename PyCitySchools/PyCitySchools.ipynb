### Summary

This project explored various educational metrics and performance indicators for the PyCity School District using data which includes information on student performance in both math and reading. Data was also provided about the schools in the District: size, budget, and type (charter or district). This analysis seeks to gain insights about individual school performance and identify trends that can inform decision-making to improve student outcomes. 

##### Metrics and Indicators:
The key metrics and performance indicators that have been analyzed in each school and/or the District includes:

- The total number of students 

- The average math and reading score

- The total number of students passing math and/or reading

- The District, school, and individual student budget

- The top and bottom performing schools based on test scores

- Math and reading scores by grade

- Scores by school spending

- Scores by school size

- Scores by school type 

- District summary

- School summaries such as by type, size, spending, and test scores


##### Methodology:
The methodology includes:

- Descriptive statistics: Calculating means, percentages, sum and other summary statistics for the key metrics.

- Comparative analysis: Comparing school performance based on different metrics (e.g. scores by school type, size, spending, and performamce in math and reading). 

- Correlation analysis: Examining potential relationships between metrics (e.g., correlation between school type and student performance).

- Data visualization: Analyses are presented as tables


### Comparisons:

##### Impact of School Size on Academic Performance:
The analysis suggests that smaller schools, with fewer than 2000 students, tend to have higher performance in math and/or reading scores  compared to larger-sized schools. It could be that smaller schools may provide a more conducive learning environment with a lower teacher to student ratio that allows for more individual attention and support to students. As school size increases, there might be challenges in maintaining the same level of individualized attention, or other factors such as larger staff to mamage, that could potentially lead to lower academic performance.

##### Differences Between Charter and District Schools:
The analysis also indicates that charter schools, on average, have higher passing rates in both math and reading compared to district schools. The charter school model and culture might be more effective in promoting academic achievement and supporting students to meet proficiency levels. District schools may have systems and practices that relate to lower academics. 

### Conclusion 
School size and type are likely key factors that influence academic performance, with smaller schools and charter schools showing higher average scores and passing rates. However, further investigation is needed to understand the underlying reasons behind these differences and to make informed decisions for school improvement and resource allocation.

```python
# Dependencies and Setup
import pandas as pd
```

```python
# Read School and Student Data Files and store into Pandas DataFrames
school_data = pd.read_csv("Resources/schools_complete.csv") 
student_data = pd.read_csv("Resources/students_complete.csv")  

# Combine into a single dataset 
complete_data = pd.merge(student_data, school_data, how="left", on="school_name")

# Preview the data  
complete_data.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = complete_data["school_name"].nunique()

# Display the result
school_count 
```

```python
# Calculate the total number of students
student_count = complete_data["student_name"].count()

# Display the result
student_count    
```

```python
# Calculate the total budget 
total_budget = school_data["budget"].sum()

# Display the result    
total_budget
```

```python
# Calculate the average (mean) math score
ave_mat_sco = complete_data["math_score"].mean()

# Display the result    
ave_mat_sco
```

```python
# Calculate the average (mean) reading score
ave_read_sco = complete_data["reading_score"].mean()    

# Display the result   
ave_read_sco
```

```python
# Calculate the percentage of students who passed math (math scores greather than or equal to 70)
total_pass_mat = complete_data.loc[student_data["math_score"] >=70] ["student_name"].count()
perc_pass_mat = total_pass_mat/student_count * 100

# Display the result    
perc_pass_mat
```

```python
# Calculate the percentage of students who passed reading (reading scores greather than or equal to 70)
total_pass_read = complete_data.loc[student_data["reading_score"] >=70] ["student_name"].count()   
perc_pass_read = total_pass_read/student_count * 100  

# Display the result
perc_pass_read
```

```python
# Calculate the percentage of students that passed math and reading with scores greater than or equal to 70 
total_pass_mat_read = complete_data.loc[(student_data["math_score"] >=70) & (student_data["reading_score"] >=70)] ["student_name"].count()  
perc_pass_mat_read = total_pass_mat_read / student_count * 100   

# Display the result
perc_pass_mat_read
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = {
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [ave_mat_sco], 
    "Average Reading Score": [ave_read_sco],
    "% Passing Math": [perc_pass_mat],
    "% Passing Reading": [perc_pass_read],
    "% Overall Passing": [perc_pass_mat_read]   }

# Create the district_summary DataFrame
district_summary = pd.DataFrame(district_summary)

# Format display
district_summary["Total Students"] = district_summary["Total Students"].apply(lambda x: "{:,}".format(x))
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary  

```python
#List schools and types to match the school_count value and avoid errors in upcoming dataframe(s) that include school_types
school_types = school_data.set_index(["school_name"])["type"]

#Display the result
school_types
```

```python
# Calculate the total student count per school
per_sch_count = complete_data.groupby("school_name").count()["student_name"]

# Display the result   
per_sch_count
```

```python
# Calculate the total budget per school
per_sch_budg = complete_data.groupby("school_name")["budget"].mean()

# Calculate the per capita spending per school
per_sch_cap = per_sch_budg / per_sch_count

# Preview the results
print(per_sch_budg.head()) 
print(per_sch_cap.head())
```

```python
# Calculate the average test scores per school
per_sch_mat = complete_data.groupby("school_name")["math_score"].mean()
per_sch_read = complete_data.groupby("school_name")["reading_score"].mean() 

# Preview the results
print(per_sch_mat.head())
print(per_sch_read.head())
```

```python
# Calculate the number of students per school with math scores of 70 or higher
stud_pass_mat =complete_data.loc[complete_data["math_score"] >=70]
sch_stud_pass_mat = stud_pass_mat.groupby("school_name")["student_name"].count()

# Preview the result
sch_stud_pass_mat.head()
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
stud_pass_read = complete_data.loc[complete_data["reading_score"] >=70]
sch_stud_pass_read = stud_pass_read.groupby("school_name")["student_name"].count()   

# Preview the result
sch_stud_pass_read.head()  
```

```python
# Calculate the number of students per school that passed both math and reading with scores of 70 or higher
stud_pass_mat_read = complete_data[(complete_data["reading_score"] >= 70) & (complete_data["math_score"] >= 70)]
sch_stud_pass_mat_read = stud_pass_mat_read.groupby(["school_name"]).size()

# Display the result
sch_stud_pass_mat_read.head() 
```

```python
# Calculate the per school passing rates
per_sch_pass_mat = sch_stud_pass_mat / per_sch_count * 100
per_sch_pass_read = sch_stud_pass_read / per_sch_count * 100
overall_pass_rate = sch_stud_pass_mat_read / per_sch_count * 100

#Display the results
print(per_sch_pass_mat.head())
print(per_sch_pass_read.head())    
print(overall_pass_rate.head())  
```

```python

# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({
    "School Type": school_types.groupby(school_types.index).first(),
    "Total Students": per_sch_count.groupby(per_sch_count.index).sum(),
    "Total School Budget": per_sch_budg.groupby(per_sch_budg.index).first(),
    "Per Student Budget": per_sch_cap.groupby(per_sch_cap.index).first(),
    "Average Math Score": per_sch_mat.groupby(per_sch_mat.index).mean(),
    "Average Reading Score": per_sch_read.groupby(per_sch_read.index).mean(),
    "% Passing Math": per_sch_pass_mat.groupby(per_sch_pass_mat.index).mean(),
    "% Passing Reading": per_sch_pass_read.groupby(per_sch_pass_read.index).mean(),
    "% Overall Passing": overall_pass_rate.groupby(overall_pass_rate.index).mean()})

# Format display
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

per_school_summary.index.name = None
# Display the per_school_summary DataFrame
per_school_summary                                 
```

## Highest-Performing Schools (by % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.
top_schools = per_school_summary.sort_values("% Overall Passing", ascending=False)

# Display the top 5 schools  
top_schools.head()
```

## Lowest Performing Schools (By % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.
bottom_schools = per_school_summary.sort_values("% Overall Passing", ascending=True)

# Format display 
bottom_schools.index.name = None

#Display the bottom 5 schools   
bottom_schools.head()   
```

## Scores by School Spending


```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spend_df = per_school_summary.copy()
```

```python
# Convert "Per Student Budget" to float (remove $ sign) and then use `pd.cut` to categorize spending based on the bins.
school_spend_df["Spending Ranges (Per Student)"] = pd.cut(school_spend_df["Per Student Budget"].replace('[\$,)]','', regex=True).astype('float'), spending_bins, labels=labels) 

# Display the data frame    
school_spend_df
```

```python
# Group by "Spending Ranges (Per Student)" and calculate the mean for each metric to create spending summary dataframe  
spending_summary = school_spend_df.groupby(["Spending Ranges (Per Student)"]).agg({
    "Average Math Score": "mean",
    "Average Reading Score": "mean",
    "% Passing Math": "mean",
    "% Passing Reading": "mean",
    "% Overall Passing": "mean" })

# Display spending summary dataframe
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], size_bins, labels=labels)

# Format display
per_school_summary.index.name = None

# Display DataFrame 
per_school_summary 
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
size_summary = per_school_summary.groupby("School Size").agg({
    "Average Math Score": "mean",
    "Average Reading Score": "mean",
    "% Passing Math": "mean",
    "% Passing Reading": "mean",
    "% Overall Passing": "mean"})

# Display the DataFrame  
size_summary
```
